```python
data2=pd.read_csv(r'C:\Users\kowsar\cleanfile.csv')
```

```python
data2['genres']=data2['genres'].str.split(', ')

data2['genres']
```

```python
data2.head()
```

```python
data_copy=data2.copy()
# Iterate through the rows of the DataFrame
for index, row in data_copy.iterrows():
    # Iterate through the genres in the 'genre' column
    for genre in row['genres']:
        # Update the 'data_copy' DataFrame based on the presence of genres
        data_copy.at[index, genre] = 1
data_copy.fillna(0, inplace=True) 
```

```python
data3=data_copy.drop('genres', axis=1)
data3
```

```python
data4=data3.drop('popularity', axis=1)
data5=data4.drop('vote_average', axis=1)
data6=data5.drop('vote_count', axis=1)
data7=data6.drop('overview', axis=1)
data8=data7.drop('release_date', axis=1)
data9=data8.drop('original_language', axis=1)
data9
```

```python
#Creating user profile
user_profile=[{'title':'Red Jasmine', 'rating':80},
             {'title':'Annai Bhoomi 3D', 'rating':90},
             {'title':'1-Jan', 'rating':50},
             {'title':'3 Idiots', 'rating':30},
             {'title':'Looop Lapeta', 'rating':20}]
df=pd.DataFrame(user_profile)
df
```

```python
title=[]
for y in df['title']:
    title.append(y)
print(title)    
```

```python
# the genre of move that my user watched
user_move_matrix=data9[data9['title'].isin(title)]
user_move_matrix
```

```python
user_matrix=user_move_matrix.drop('title', axis=1)
user_matrix
```

```python
q1=user_matrix.reset_index(drop=True)
q1
```

```python
#weighting the quality of move
weighted_genre_matrix=q1.transpose().dot(df['rating'])
weighted_genre_matrix
```

```python
#recommendation table: multiplying user profile into all course skills
recommendation=((data9*weighted_genre_matrix).sum(axis=1))/(weighted_genre_matrix.sum())
rr=recommendation.sort_values(ascending=False)
rr
```
